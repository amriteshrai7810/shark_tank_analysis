```python
# Import the libraries

import pandas as pd
```

```python
df = pd.read_csv('Shark_Tank_India_uncleaned.csv')
```

```python
# Show all columns
pd.set_option('display.max_columns', None)
```

```python
df.shape
```

```python
df.head()
```

```python
# Dealing with the int missing values

df['Male_Presenters'].fillna(0, inplace=True)
df['Female_Presenters'].fillna(0, inplace=True)
df['Transgender_Presenters'].fillna(0, inplace=True)
df['Couple_Presenters'].fillna(0, inplace=True)
df['Number_of_Sharks_in_Deal'].fillna(0, inplace=True)
```

```python
# Change Data Type

df['Season_Number'] = df['Season_Number'].astype('int8') # Optimized data types to reduce file size
df['Episode_Number'] = df['Episode_Number'].astype('int16')
df['Pitch_Number'] = df['Pitch_Number'].astype('int8')
df['Season_Start'] = df['Season_Start'].astype('datetime64')
df['Season_End'] = df['Season_End'].astype('datetime64')
df['Original_Air_Date'] = df['Original_Air_Date'].astype('datetime64')
df['Number_of_Presenters'] = df['Number_of_Presenters'].astype('int8')
df['Male_Presenters'] = df['Male_Presenters'].astype('int8')
df['Female_Presenters'] = df['Female_Presenters'].astype('int8')
df['Transgender_Presenters'] = df['Transgender_Presenters'].astype('int8')
df['Couple_Presenters'] = df['Couple_Presenters'].astype('int8')
df['Number_of_Sharks_in_Deal'] = df['Number_of_Sharks_in_Deal'].astype('int8')
```

```python
df.info()
```

```python
# Check null values
df.isna().sum()
```

```python
# Drop columns we don't need 
df.drop(columns=['Company_Website','Original_Air_Date','Episode_Title','Gross_Margin','Net_Margin','EBITDA','Cash_Burn','SKUs','Has_Patents','Bootstrapped'], inplace=True)
```

```python
# Filling na values
df['Started_in'] = df['Started_in'].fillna("Not Mentioned")
df['Pitchers_City'] = df['Pitchers_City'].fillna('Not Mentioned')
df['Pitchers_State'] = df['Pitchers_State'].fillna('Not Mentioned')
df['Accepted_Offer'] = df['Accepted_Offer'].fillna("No Offer Received.")
df['Accepted_Offer'] = df['Accepted_Offer'].fillna("No Offer Received")
df['Total_Deal_Amount(in_lakhs)'] = df['Total_Deal_Amount(in_lakhs)'].fillna(0)
df['Total_Deal_Equity(%)'] = df['Total_Deal_Equity(%)'].fillna(0)
df['Number_of_Sharks_in_Deal'] = df['Number_of_Sharks_in_Deal'].fillna(0)
df['Namita_Investment_Amount(in lakhs)'] = df['Namita_Investment_Amount(in lakhs)'].fillna(0)
df['Vineeta_Investment_Amount(in_lakhs)'] = df['Vineeta_Investment_Amount(in_lakhs)'].fillna(0)
df['Anupam_Investment_Amount(in_lakhs)'] = df['Anupam_Investment_Amount(in_lakhs)'].fillna(0)
df['Aman_Investment_Amount(in_lakhs)'] = df['Aman_Investment_Amount(in_lakhs)'].fillna(0)
df['Peyush_Investment_Amount((in_lakhs)'] = df['Peyush_Investment_Amount((in_lakhs)'].fillna(0)
df['Amit_Investment_Amount(in_lakhs)'] = df['Amit_Investment_Amount(in_lakhs)'].fillna(0)
df['Ashneer_Investment_Amount'] = df['Ashneer_Investment_Amount'].fillna(0)
df['Namita_Present'] = df['Namita_Present'].fillna('No')
df['Vineeta_Present'] = df['Vineeta_Present'].fillna('No')
df['Anupam_Present'] = df['Anupam_Present'].fillna('No')
df['Aman_Present'] = df['Aman_Present'].fillna('No')
df['Peyush_Present'] = df['Peyush_Present'].fillna('No')
df['Amit_Present'] = df['Amit_Present'].fillna('No')
df['Ashneer_Present'] = df['Ashneer_Present'].fillna('No')
```

```python
# Filling null values Yearly_Revenue(in_lakhs) and Monthly_Sales(in_lakhs)

df['Yearly_Revenue(in_lakhs)'] = df['Yearly_Revenue(in_lakhs)'].fillna(-1)
df['Monthly_Sales(in_lakhs)'] = df['Monthly_Sales(in_lakhs)'].fillna(-1)
```

```python
# Cleaning anchor, industry business description

df['Anchor'] = df['Anchor'].str.replace('0',' ')
df['Industry'] = df['Industry'].str.replace('0',' ')
df['Business_Description'] = df['Business_Description'].str.replace('0',' ')
```

```python
# Renaming columns with '()' as it can cause issue in SQL

# Get the names with brackets in them
col_brackets = []
col_no_brackets = []

for i in df.columns:
    if '(' in i:
        col_brackets.append(i)
    
        
for i in col_brackets:
    x = i.replace(')', '')
    col_no_brackets.append(x.replace('(', '_'))
    
for i in range(0, len(col_brackets)):
    df.rename(columns={col_brackets[i]:col_no_brackets[i]}, inplace=True)
```
